Resizing the image using Pillow library

```python
from PIL import Image
```

```python
img=Image.open('/content/th?id=OIP.6_bAQdISM03qiib1a2b1eAHaEo&pid=Api&P=0&h=220.2')
img_resized=img.resize((200,200))
```

```python
img_resized.save('Horse_img_resized.jpg')
```

```python
# displaying the image from numpy array
img_res=mpimg.imread('/content/Horse_img_resized.jpg')
img_res_plot=plt.imshow(img_res)
plt.show()
```

Converting RGB images to Grayscale image using OpenCV(Computer Visions)

```python
# importing OpenCV library
import cv2
```

```python
img=cv2.imread('/content/th?id=OIP.6_bAQdISM03qiib1a2b1eAHaEo&pid=Api&P=0&h=220.2')
```

```python
type(img)
```

```python
img.shape
```

```python
grayscale_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
type(grayscale_img)
```

```python
grayscale_img.shape
```
